```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('text_emotion.csv')
```

```python
df
```

```python
df.drop(['tweet_id', 'author'], axis =1, inplace = True)
```

```python
df = df.loc[(df['sentiment']=='enthusiasm')| (df['sentiment']=='sadness')]
```

```python
df.loc[df['sentiment']=='enthusiasm', 'sentiment'] =1
df.loc[df['sentiment']=='sadness', 'sentiment'] = 0
```

```python
x = df['content']
y = df['sentiment']
```

```python
#x = np.asarray(x).astype(np.float32)
y = np.asarray(y).astype(np.float32)
```

```python
#X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
#creating a split (train/test)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=45)
```

```python
from tensorflow.keras.models import Sequential 
sentiment = Sequential()
```

```python
from tensorflow.keras.layers import Embedding
max_size = 20000
max_length = 130
sentiment.add(Embedding(input_dim = max_size, 
                    output_dim = 10,
                    input_shape = (max_length - 1,)))
```

```python
from tensorflow.keras.layers import LSTM 
sentiment.add(LSTM(units = 128))
```

```python
sentiment.add(LSTM(units = 64))
```

```python
from tensorflow.keras.layers import Dense
sentiment.add(Dense(units=100, activation='relu'))

#signlangmodel.add(Dense(units=10, activation='softmax'))
```

```python
from tensorflow.keras.layers import Dropout, Dense 
sentiment.add(Dropout(rate=0.5))
```

```python
sentiment.add(Dense(units=5, activation ='softmax'))
```

```python
sentiment.summary()
```

```python
sentiment.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])
sentiment.fit(x, y, epochs=5)
```

```python
scores = sentiment.evaluate(x, y, verbose=0)
print("%s: %.2f%%" % (sentiment.metrics_names[1], scores[1]*100))
```
